```python
import numpy as np
import math

# -------REGULA-DELTA--------

#DECLARARE  SI INITIALIZARE DATE 
y=np.array([[45,85,-1],[50,43,-1],[40,80,-1],[55,42,-1],[200,43,-1],[48,40,-1],[195,41,-1],[43,87,-1],[190,40,-1]])


col_norm_x = (y[:,0]-min(y[:,0])) / (max(y[:,0])-min(y[:,0]))
col_norm_y = (y[:,1]-min(y[:,1])) / (max(y[:,1])-min(y[:,1]))
y_norm=[]

# print(col_norm_x,col_norm_y)
for i in range(len(y)):
    y_norm.append([col_norm_x[i],col_norm_y[i],-1])

print(y_norm)
#iesiri dorite
d1 = np.array([1,-1,-1]) #scaun
d2 = np.array([-1,1,-1]) #mese
d3 = np.array([1,-1,-1]) 
d4 = np.array([-1,1,-1])
d5 = np.array([-1,-1,1]) #paturi
d6 = np.array([-1,1,-1])
d7 = np.array([-1,-1,1])
d8 = np.array([1,-1,-1])
d9 = np.array([-1,-1,1])
d = [d1,d2,d3,d4,d5,d6,d7,d8,d9]
w=[]
# w=np.array([[1,-1,-1],[1,1,-1],[-1,1,-1 ]])
index=0
while index<3 :
    w.append(np.random.uniform(-1,1,size = 3))
    index =  index +1

M = 9 # nr de intrari
c = 0.2 # constanta de instruire
e_max = 9
# eroarea maxima
e = 0
o=[]      
net=[0,0,0]
for i in range(M):
     o.append([[],[],[]])

#functie pentru calculul functie de activare bipolara
def f (net):
    return 2/(1+np.exp(-net))-1

#se calculeaza iesirile fiecarui perceptron
while(1):
    e = 0
    for p in range(M):
        for j in range(3):
            net=w[j].dot(y_norm[p])
#             print(net)
            o[p][j]=f(net)
#             print("IESIRILE FIECARUI PERCEPTRON\n",o)
            for i in range(3):
                w[j][i]= w[j][i]+c*np.multiply((d[p][j]-o[p][j])*(1-o[p][j]*o[p][j]),y_norm[p][j])
        e=e+np.sum(np.power(d[p][j]-o[p][j],2))
    print("Iesiri  \n",o)
    print("Eroare",e)
    print("-----------------------------------------------")
    if e<e_max: break
```

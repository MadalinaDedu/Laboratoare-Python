```python
import numpy as np
import math
import random

# DECLARARE SI INITIALIZARE DATE
p1 = [45,85]
p2 = [50,43]
p3 = [40,80]
p4 = [55,42]
p5 = [200,43]
p6 = [48,40]
p7 = [195,41]
p8 = [43,87]
p9 = [190,40]

p = [p1,p2,p3,p4,p5,p6,p7,p8,p9]
centroizi_initiali = random.sample(p,3)
centroizi_preced = []
cluster = []
print("Puncte:\n",p,"\n")
print("Centroizi initiali:\n",centroizi_initiali,"\n")

def dist_euclidiana(a,b):
    dist_e = math.sqrt((a[0]-b[0])*(a[0]-b[0]) + (a[1]-b[1])*(a[1]-b[1]))
    return dist_e

while(1):
    for i in range(len(p)):
        distante=[]
        for j in range(len(centroizi_initiali)):
    #         print(p[i])
            dist = dist_euclidiana(p[i],centroizi_initiali[j]) 
            distante.append(dist)
    #         print(distante)
        dist_min=np.argmin(distante)
        distante.clear()
        cluster.append(dist_min)

    centroizi_noi = []
    centru_greutate_nou = []

    for k in range(len(centroizi_initiali)):
        print("Punctele centroidului ", k ,"sunt")
        sum_x = 0
        sum_y = 0
        nr = 0
        for j in range(len(p)):
            if cluster[j]==k :
                sum_x = sum_x+p[j][0]
                sum_y = sum_y+p[j][1]
                nr = nr+1
                print(p[j])

        x = sum_x/nr
        y = sum_y/nr
        centru_greutate_nou.append(x)
        centru_greutate_nou.append(y)
    centroizi_noi.append(centru_greutate_nou)
        
    if np.array_equal(centroizi_noi,centroizi_preced) :
        break
    print("Centroizi noi",centroizi_noi)
    centroizi_preced = centroizi_noi
    cluster.clear()
```

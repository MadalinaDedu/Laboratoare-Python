```python
# regresie liniara

import csv
import random

# citirea datelor din fisier
myList = []
with open(r'C:\Users\user\Salary_Data.csv') as csvfile:
    readCSV = csv.reader(csvfile)
    for row in readCSV:
        if len(row) == 2:
             myList.append([float(row[0]), float(row[1])])
                
# generarea random a valorilor initiale ale ponderilor
w1=random.randint(1, 50)
w2=random.randint(1, 50)

c = 0.01
index = 0
n = len(myList)
suma = 0

#bucla repetitiva se executa pana cand una din variabile este stabila
while (1) : 
    df_w1 = 0 
    df_w2 = 0
    for i in range(n):
        df_w1 = df_w1 + ( myList[i][1]-(w1*myList[i][0]+w2))*myList[i][0]
        df_w2 = df_w2 + myList[i][1]-(w1*myList[i][0]+w2)
    df_w1 = -df_w1/n
    df_w2 = -df_w2/n

    w_n1 = w1 - c*df_w1
    w_n2 = w2 - c*df_w2
    index = index + 1
    
    print("Parametrii dreptei de regresie dupa -> ", index, "iteratii este:")
    print("w1 =", w_n1,"w2 = ",w_n2,"\n")
  

    if abs(w_n1-w1) <= 0.00001 and abs(w_n2-w2) <= 0.00001 : 
        break
    w1 = w_n1
    w2 = w_n2
```

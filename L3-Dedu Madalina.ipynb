```python
# f(x) = 6x^2-12x+1
# f'(x) = 12x-12 = 12(x-1)
# x_n+1 ->not-> x_n1 = x_n-12*c*(x-1)

#  doua valori luate aleator pentru constanta c si x_n1
x_n1 = 1
c = 0.01
x_n = 3
index = 0

while x_n1 > 0.9 : 
    x_n1 = x_n - 12*c*(x_n-1)
    index = index + 1
    print("Minimiul functie este dupa ", index, "iteratii este: ", x_n1)
    if abs(x_n-x_n1) <= 0.00001: 
        break

    x_n = x_n1
    
   
```

```python
# g(x,y) = x^2+2y^2
#g'(x) = 2x
#g'(y) = 4y
#Minimiul functie este dupa  1458 iteratii este: x=   1.0144508956538812e-141  y= 5e-324
#Minimiul functie este dupa  1459 iteratii este: x=   8.11560716523105e-142  y= 5e-324
    
x_n1 = 1
x_n = 2
y_n1 = 1
y_n = 2
c = 0.1
index = 0

#bucla repetitiva se executa pana cand una din variabile este stabila
while x_n1 > 0 and y_n1 > 0 : 
    x_n1 = x_n - 2*c*x_n
    y_n1 = y_n - 4*c*y_n
    index = index + 1
    print("Minimiul functie este dupa ", index, "iteratii este: x=  ", x_n1," y=",y_n1)
    if abs(x_n-x_n1) <= 0.00001 and abs(y_n-y_n1) <= 0.00001 : 
        break

    x_n = x_n1
    y_n = y_n1
```

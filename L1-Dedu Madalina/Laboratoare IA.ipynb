```python
import csv
myList = []
with open('c:\\Users\\user\\Desktop\\iris.data') as csvfile:
    readCSV = csv.reader(csvfile, delimiter=',')
    for row in readCSV:
        if len(row) == 5:
             myList.append([float(row[0]), float(row[1]), float(row[2]), float(row[3])])

```

```python
import numpy as np
npArray = np.array(myList)
x=0
for x in range (len(npArray[0])) :
    col_min = np.min(npArray[:,x])
    col_max = np.max(npArray[:,x])
    col_medie = np.average(npArray[:,x])
    col_mediana = np.median(npArray[:,x])
    col_norm = (npArray[:,x]-min(npArray[:,x])) / (max(npArray[:,x])-min(npArray[:,x]))

    print("Col min " , x , "este ", col_min)
    print("Col col_max " , x , "este ", col_max)
    print("Col col_medie " , x , "este ", col_medie)
    print("Col col_mediana " , x , "este ", col_mediana)
    print("Col col_norm " , x , "   este ","\n", col_norm)
    print("\n")
    print("\n")
    

```

```python
pondere_array = []  
row_pondere = []
suma = 0
pondere = [0.2,1.1,-0.9,1]

for j in range (len(npArray)) :
    for i in range(len(pondere)):
        row_pondere = pondere[i]*npArray[j,i]
        suma = suma + row_pondere
      
    pondere_array.append(suma)
    suma = 0   
print (pondere_array)



```
